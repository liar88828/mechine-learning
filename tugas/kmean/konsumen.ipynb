```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# meyiapkan data dan memangil data set
dataset = pd.read_csv('./../../data/konsumen.csv')
dataset.keys()
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# konverksi ke data array
x = np.asarray(dataset)
print(x)
```

```python
# menampilkan data dalam bentuk scatter plot
plt.scatter(
    x[:, 0],
    x[:, 1],
    label='True Position')
plt.xlabel('Gaji')
plt.ylabel('Pengeluaran')
plt.title('Grafik Penyebaran Data Konsumen')
plt.show()
```

```python
# Mengakrtifkan K mean dengan jumlah K=2
kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
```

```python
# menampilkan nilai centroid yang di generate oleh algoritma 
print(kmeans.cluster_centers_)
```

```python
#plot Data point
plt.scatter(
    x[:, 0],
    x[:, 1],
    c=kmeans.labels_,
    cmap='rainbow'
)
plt.xlabel('Gaji')
plt.ylabel('Pengeluaran')
plt.title('Grafik Penyebaran Data Konsumen')
plt.show()
```

```python
#plot Data point
# menvisualisasikan hasil klastering dengan centeroid dari masing masing klaster
plt.scatter(
    x[:, 0],
    x[:, 1],
    c=kmeans.labels_,
    cmap='rainbow'
)
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    color='black'
)
plt.xlabel('Gaji')
plt.ylabel('Pengeluaran')
plt.title('Grafik Penyebaran Data Gaji dan pengeluaran Konsumen')
plt.show()
```
